```python
#Categorize the people by their age like children, adult, citizen, senior citizen etc.,
age= int(input("age:"))
if(age < 18) :
    print("Children")
elif(age < 35) :
    print("Adult")
elif(age < 60) : 
    print("Citizen")
else : print("Senior Citizen")     
```

```python
#Find whether the given number is positive or negative
num=int(input("Enter any number : "))
if (num < 0 ) : 
   print ("Number is negative")
elif(num == 0) : 
   print ("number is zero")
else : print ("Number is positive")
```
