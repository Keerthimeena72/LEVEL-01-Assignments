```python
#Print odd numbers in the list
list = (20,10,16,19,25,1,276,188)
for num in list:
    if num%2 != 0 :
       print(num, "is odd")
```

```python
#Print even numbers in the list 
list = (20,10,16,19,25,1,276,188)
for num in list:
    if num%2 == 0 :
       print(num, "is even")
```
